```python
import pandas as pd
import string
import re
```

```python
df = pd.read_csv('/content/ccibeekeoc42-dollyhhrlhf-yoruba.csv')
df.head()
```

```python
# Check for null values

print(df.isnull().sum())
```

```python
def clean_text(text):
    text = re.sub(r'[^\w\sÀ-ÿ]', '', text)  # Remove punctuation but keep Yoruba characters
    text = re.sub(r'\s+', ' ', text)  # Normalize spaces
    return text.strip()
```

```python
df['clean_Response_Yoruba'] = df['Response_Yoruba'].apply(clean_text)
df[['Response_Yoruba', 'clean_Response_Yoruba']].head()
```

```python
# Word count per entry
df['Word_Count'] = df['Response_Yoruba'].apply(lambda x: len(x.split()))

# Show stats
total_words = df['Word_Count'].sum()

print("Total words:", total_words)
```

```python
# Sentence count per entry (naively splitting on full stop or question mark)
df['Sentence_Count'] = df['Response_Yoruba'].apply(lambda x: len(re.findall(r'[.!?]', x)))

# Show stats
total_sentences = df['Sentence_Count'].sum()

print("Total sentences:", total_sentences)
```

```python
print(df[['Response_Yoruba', 'Word_Count', 'Sentence_Count']].head())
```

```python
from collections import Counter
```

```python
# Join and split all words
all_words = ' '.join(df['Response_Yoruba']).split()
word_freq = Counter(all_words)

# Get the top 20 most common words
common_words = word_freq.most_common(20)

# Create a DataFrame for better display
freq_df = pd.DataFrame(common_words, columns=['Word', 'Frequency'])
print(freq_df)
```
